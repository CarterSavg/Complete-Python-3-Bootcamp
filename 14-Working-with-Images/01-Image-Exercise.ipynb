```python
from PIL import Image

sheet = Image.open("word_matrix.png")

sheet.putalpha(128)
```

```python
mask = Image.open("mask.png")
mask = mask.resize(size=(1015, 559))
mask.putalpha(128)
```

```python
sheet.paste(im=mask, box=(0,0), mask=sheet)
sheet
```
